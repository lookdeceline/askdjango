# 1. HTML Form 
- 웹페이지에서는 form 태그를 통해 데이터를 전송
    - ex) login form, comment form
- 하나의 form 태그는 하나 이상의 widget을 갖는다 



## 1-1. HTML Form 태크 필수 속성
#### 1. action: 요청을 보낼 주소
#### 2. method: GET or POST
    - "GET": 주로 데이터 조회 요청 시
        - url주소 뒤에 "?= ( 유저로부터 받은 GET요청 )"을 붙여서 보냄.
        - header만 간다. body가 없다. 
        - 인코딩된 데이터 묶음을 header로만 보낸다.
        - GET 인자: 
    - "POST": 파괴적인 액션(생성/수정/삭제)에 대한 요청
        - url주소 뒤에 붙이지 않고 body 로 보낸다.
#### 3. enctype: request.POST 요청시에만 유효
    - urlencode
    - multipart
