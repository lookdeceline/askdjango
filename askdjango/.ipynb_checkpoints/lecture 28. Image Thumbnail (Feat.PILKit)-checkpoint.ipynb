# Pilkit

## 1. requests를 이용하여 이미지 다운받기

```python
img_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/d7/Carina_Dwarf_Galaxy.jpg/1200px-Carina_Dwarf_Galaxy.jpg"
```

```python
import requests
img_binary = requests.get(img_url).content
with open("galaxy.jpg", "wb") as f:
    f.write(img_binary)
```

```python
from PIL import Image
galaxy_img = Image.open("galaxy.jpg")
```

```python
galaxy_img
```

## 2. Thumbnail

```python
from PIL import Image
from pilkit.processors import Thumbnail

processor = Thumbnail(width=300, height=300)
```

```python
thumb_img = processor.process(galaxy_img)
```

```python
thumb_img
```

```python
thumb_img.save("thumb-300x300.png")
thumb_rgb = thumb_img.convert('RGB')
for quality in [100, 80, 60, 40, 20]:
    thumb_rgb.save("thumb-300x300-{}.jpg".format(quality), quality=quality)
```
