# what is Filter

## Django Template Filter
- 템플릿 변수값 변환을 위한 함수. 다수 필터 함수 연결 가능 
    - {{ var|filter1 }}, {{ var|filter2:인자 }}, {{ var|filter3:인자|filter4 }}

### ex: date time 

```python
# in django
from django.utils import timezone
timezone.now()
timezone.now().strftime("%Y-%m-%d %H:%M:%S")
```

### safe filter
- HTML Escaping이 수행되지 ㅏㅇㄶ도록 문자열을 SafeString으로 변환.
- autoescaping 이 off로 지정될 경우, 이 필터는 작동하지 않음

### slice filter
- same as Python slicing
- {{ some_list|slice:"1:10:2"}}

### striptags filter
- HTML 태그를 모두 제거
- non valid HTML일 경우 제거가 되지 않을 수 있음.
- 더 견고한 제거는 bleach 파이선 라이브러리의 clean 메소드 출력

- value = "<b>Joel</b> Nbutton>is</button> a <span>slug</span>"
- {{ value | striptags }} 
- ==> "Joel is a slug"

### truncatechars filter
- 문자열을 지정 글자갯수까지 줄이고, 줄여진 경우 끝에 "..."를 추가

### truncatechars_html filter
- html태그를 보호하면서 문자열을 지정 글자 개수까지 줄임

```python
value1 = "Joel is a slug"
value2 = "<p>Jeol is a slug</p>"
{{ value1|truncatechars:9 }} => "Joel i..."
{{ value2|truncatechars_html:9}} => "<p>Joel i...</p>"
```
