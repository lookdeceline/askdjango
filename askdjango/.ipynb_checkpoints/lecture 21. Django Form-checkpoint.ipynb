# Django Form
- front end와 인터페이싱을 위해 만드는 클래스
- 주요 커스텀 form class:
    - 입력폼 HTML 생성: .as_table(), .as_p(), .as_ul()
    - 입력폼 값 검정 / 값 변환
    - 검증을 통과한 값들을 사전타입으로 제공: cleaned_data
    
    
- Models Fields vs Form Fields
    - Model Fields: Database Field들을 파이썬 클래스화
    - Form Fields: HTML Form Field들을 파이썬 클래스화
