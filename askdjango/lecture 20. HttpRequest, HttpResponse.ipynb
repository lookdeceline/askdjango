# HttpRequest & HttpResponse
### 모든 view 함수는 HttpRequest(request)를 첫번째 인자로 받으며, HttpResponse 인스턴스를 리턴한다.

# I. HttpRequest
- 클라이언트로부터의 모든 요청 내용을 담고 있음.
- 매 요청 시마다 **view 함수의 첫번째 인자**로 전달


#### HttpRequest objects에서 폼 처리 관련 속성들
- request.method: 요청의 종류. 무조건 GET or POST
- request.GET: GET인자. QueryDict type.
- request.POST: POST인자. QueryDict type.
- request.FILES: POST업로드 파일인자. MultiValueDict type.


###### MultiValueDict vs Query Dict
**MultiValueDict:**
- 동일 key의 다수 value를 지원
- 

**QueryDict(MultiValueDict)**
- immutable

# II. HttpResponse
- 뷰 함수의 리턴값으로서 HttpResopnse의 인스턴스를 기대.
    - view ==> client 사이의 Middleware들은 view 함수에서 받은 값이 response type이려니 기대하고, response type을 처리하기 때문에.
- 여러 helper함수를 통해 손쉽게 HttpResponse 인스턴스를 생성.
    - ex) django.shortcuts.**render**
    - django.shorcuts.**redirect**
    - django.http.**JsonResponse** 등
